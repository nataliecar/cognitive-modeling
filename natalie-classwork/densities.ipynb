```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1D case
x = np.random.normal(loc=0, scale=1, size=100000)
```

```python
fig,ax = plt.subplots(1,1,figsize=(6,4))
sns.histplot(x, ax=ax, bins=50, color='skyblue')
```

```python
x = np.random.multivariate_normal([0.,0.], np.eye(2), size=10000)
print(x.shape)
```

```python
f,ax = plt.subplots(1,1, figsize=(5,5))
ax.scatter(x[:,0], x[:,1], color='skyblue')
```

```python
cov = np.array([[1,0.5], [0.5,1]])
print(cov)
x = np.random.multivariate_normal([3.,0.], cov, size=10000)
f, ax = plt.subplots(1,1, figsize=(5,5))
ax.scatter(x[:,0], x[:,1], color='skyblue')
```

```python
cov = np.array([[1,2], [2,1]])
x = np.random.multivariate_normal([0.,0.], cov, size=10000)
f, ax = plt.subplots(1,1, figsize=(5,5))
ax.scatter(x[:,0], x[:,1], color='skyblue')
```

```python
cov = np.array([[1,2], [2,1]])
x = np.random.multivariate_normal([0.,0.], cov, size=10000)
f, ax = plt.subplots(1,1, figsize=(5,5))
# Kernel Density Estimate
sns.kdeplot(x = x[:,0], y = x[:,1], color='skyblue', fill=True) # max 2-D vectors
```

```python
sns.jointplot(x = x[:,0], y = x[:,1], color='skyblue')
```

```python
sns.jointplot(x = x[:,0], y = x[:,1], color='skyblue', kind='kde', joint_kws={'fill':True})
```
