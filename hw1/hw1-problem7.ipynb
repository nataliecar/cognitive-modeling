```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

Question 1:

```python
N = 10000 #number of simulations
prior = np.random.rand(N) #random prior probabilities, P(x=True)
sens = 0.95 #fixed sensitivity, P(T=positive|X=true)
spec = 0.9 #fixed specificity, P(T=negative|X=false)
posteriors = (sens*prior)/(prior*sens + (1-prior)*(1-spec)) #calc posterior probabilities, P(X=true|T=positive)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 6))
ax.plot(prior, posteriors, color='#AA0000', marker='o', markersize=1, linestyle='None')
ax.set_xlabel('Prior Probability', fontsize=16)
ax.set_ylabel('Posterior Probability', fontsize=16)
ax.set_title('Posterior Probability vs. Prior Probability', fontsize=20)
ax.grid(alpha=0.2)
```
