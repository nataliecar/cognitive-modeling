```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

import warnings
warnings.filterwarnings("ignore", "is_categorical_dtype")
warnings.filterwarnings("ignore", "use_inf_as_na")
```

```python
mu_0 = 3
sigma_0 = 0.5
var = 40
n = 1000
```

```python
data = np.random.normal(2, np.sqrt(var))
y = np.mean(data)
```

```python
prior = stats.norm(mu_0, sigma_0)
mu = prior.rvs(size=n)

posterior = stats.norm((var/n * mu_0 + sigma_0**2* y)/(var/n + sigma_0**2), (n/var + 1/sigma_0**2)**-1)
```

```python
mu = np.random.normal(mu_0, sigma_0, size=n)
y = np.random.normal(mu, np.sqrt(var), size=n)

posterior = np.random.normal((var/n * mu_0 + sigma_0**2* np.mean(y))/(var/n + sigma_0**2), (n/var + 1/sigma_0**2)**-1, size=n)
```

```python
f, ax = plt.subplots(1, 1, figsize=(12, 3))
sns.histplot(mu, color='gray', ax=ax, alpha=0.5, label='Prior')
sns.histplot(posterior, color='maroon', ax=ax, label='Posterior')
ax.legend()
f.tight_layout()
```
